```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.plot(refugees['Refugees assisted by UNHCR'], color='purple')
plt.ylabel('Values')
plt.title('Refugees')

plt.subplot(1, 2, 2)
plt.scatter(x = refugees['Year'], y = refugees['Refugees assisted by UNHCR'], color='green')
plt.ylabel('Squared Values')
plt.title('Squared Series')

plt.tight_layout()
plt.show()
```

```python
# Data to play with. Twice the histograms, twice the fun.
x = x=refugees['Country or territory of asylum or residence']
y = y=refugees['Total refugees and people in refugee-like situations assisted by UNHCR']
# Override bin defaults with specific bin bounds.
# FYI `alpha` controls the opacity.
plt.hist(x, color='blue', bins=np.arange(-5, 40), alpha=.5) 
plt.hist(y, color='red', bins=np.arange(-5, 40), alpha=.5)
plt.title('Manually setting bin placement')
plt.xlabel('Random Values')

plt.show()
```
